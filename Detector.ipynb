```python
import cv2
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
test1 = cv2.imread('/home/sakshi/Downloads/Dog-with-Cat_HD_animalplanethd.com_.jpg')
test2 = cv2.imread('/home/sakshi/Downloads/Dog-with-Cat_HD_animalplanethd.com_.jpg')
```

```python
gray_img_1 = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
gray_img_2 = cv2.cvtColor(test2, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img_1, cmap='gray')
```

```python
haar_face_cascade = cv2.CascadeClassifier('/home/sakshi/Downloads/haarcascade_frontalface_default.xml')
```

```python
t1=time.time()
faces = haar_face_cascade.detectMultiScale(gray_img_1, scaleFactor=1.08, minNeighbors=8)
print(len(faces))
t2=time.time()
dt1=t2-t1
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(test1, (x, y), (x+h, y+w), (0, 255, 0), 15) #green boxes
```

```python
plt.imshow(convertToRGB(test1))
print dt1
```

```python
lbp_face_cascade = cv2.CascadeClassifier('/home/sakshi/Downloads/lbpcascade_frontalface_improved.xml')
```

```python
t3=time.time()
faces_2 = lbp_face_cascade.detectMultiScale(gray_img_2, scaleFactor=1.035, minNeighbors=5)
print(len(faces_2))
t4=time.time()
dt2=t4-t3
```

```python
for (x, y, w, h) in faces_2:
    cv2.rectangle(test2, (x, y), (x+h, y+w), (255, 0, 0), 15) #blue boxes
```

```python
plt.imshow(convertToRGB(test2))
print(dt2)
```
